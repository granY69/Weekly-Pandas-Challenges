### Q2. Delete the columns 'Unnamed: 0' and 'Id'.

```python
#axis=1 for columns
#inplace=True means apply this operation on the actual dataframe
df.drop(['Unnamed: 0', 'Id'], axis=1, inplace=True)
```

```python
df.head()
```

### Q3. Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.

```python
names = df.groupby('Name', group_keys=False).apply(lambda x : x)
```

```python
names.sort_values(['Count'], ascending=False)
```

```python
df.groupby('Name').apply(lambda x : x.sort_values(['Count'], ascending=False))
```
